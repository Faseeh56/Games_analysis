```python
data2.head()
```

```python
top_5_publisher = data2['publisher'].value_counts()[:5]
top_5_publisher
```

```python
top_5_platform = data2['platform'].value_counts()[:5]
top_5_platform
```

```python
top_5_developer = data2['developer'].value_counts()[:5]
top_5_developer
```

```python
top_5_data = data2[data2['publisher'].isin(top_5_publisher.index) & data2['platform'].isin(top_5_platform.index) & data2['developer'].isin(top_5_developer.index)]
top_5_data
```

```python
# Dogged plot
```

```python
sns.countplot(x = 'platform', hue = 'publisher', data = top_5_data)
```

```python
df_stacked_plt = pd.crosstab(index = top_5_data['platform'], columns = top_5_data['publisher'])
df_stacked_plt
```

```python
# stacked plot
```

```python
df_stacked_plt.plot(kind = 'bar', stacked = True)
```

```python
sns.boxplot(x = 'publisher', y= 'sales', data = top_5_data)
```
